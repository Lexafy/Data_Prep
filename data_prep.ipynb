# Series
Section is about series in python

```python
a = pd.Series(range(1,10))
b = pd.Series(['I', 'like', 'to', 'use', 'Python', 'and', 'pandas', 'very', 'much' ], index = range (0, 9 ))
print(a,b)
```

```python
# time series
start_date = 'Jan 1, 2020'
end_date = 'Dec 31, 2020'
idx = pd.date_range(start_date, end_date)

# literal format, f, helps to add codes to strings without loosing context. 
# Removing the literal format, f, will print out a totally different result
print(f'The length of the index is {len(idx)}')

item_sold = pd.Series(np.random.randint(25, 50, size=len(idx)), index = idx)

item_sold.plot(figsize=(8, 5))
plt.show()
```

# Dataframes
these are date in tabular format

```python
a = pd.Series(range(1,10))
b = pd.Series(['I', 'like', 'to', 'use', 'Python', 'and', 'pandas', 'very', 'much' ], index = range (0, 9 ))

df = pd.DataFrame([a,b])
df
```

```python
# To transpose the table
df.T
```

```python
# To add labels to the table

df.T.rename(columns={0: 'A', 1: 'B'})
```

```python
'''
At this stage, df is still in the horizontal view mode because technically the
Transpose done on it hasn't been saved yet.
To understand, run df below to see it output, we will then properly save it in the 
next cel
'''

df
```

```python
# Now to save the transposed df table
df = df.T.rename(columns={0: 'A', 1: 'B'})
df
```

```python
df
```

```python
# Filter:  Working with a section or just part of a dataframe
df[ df ['A'] > 5 ]
```

```python
# Freestyling with different mathematical operators in Dataframe

df[ (df ['A'] > 5) & (df ['A'] < 7 )]
```

```python
# Creatig new Column

df['DivA'] = df ['A'] - df ['A'].mean()
df
```

```python
# Adding another column filled with the length of the data in Column B calls
df['LenB'] = df['B'].apply(len)
df
```

```python
# Filtering our dataframe by number

df.iloc[:5]
```

```python
# Grouping the dataset

df.groupby(by = 'LenB').mean()
```

Column B is excluded above because the data in it cells are all strings

```python
# WHen faced with a large dataset and you wish to print a short part from the dataset
# to analyse, .head() and .tail() will printthe first five and last five rows of the 
# dataset out

df.head()
```

```python
# To specify the exact number of rows you want

df.tail(6)
```
